# Basic line plot.

```python

import matplotlib.pyplot as plt

X_Axis = [1, 2, 3, 4, 5]
Y_Axis = [10, 20, 30, 40, 50]

plt.plot(X_Axis, Y_Axis)

plt.title("Basic line plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```
